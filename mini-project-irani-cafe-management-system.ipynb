```python
# Irani Cafe Management System

# Class to represent a menu item
class MenuItem:
    def __init__(self, name, price):
        self.name = name
        self.price = price

# Class to represent the Cafe
class IraniCafe:
    def __init__(self):
        self.menu = []
        self.orders = []

    def add_menu_item(self, name, price):
        item = MenuItem(name, price)
        self.menu.append(item)

    def show_menu(self):
        print("\n---- Cafe Menu ----")
        for idx, item in enumerate(self.menu):
            print(f"{idx + 1}. {item.name} - ₹{item.price}")
        print("-------------------\n")

    def take_order(self):
        self.show_menu()
        order = []
        while True:
            choice = input("Enter item number to order (or 'done' to finish): ").strip().lower()
            if choice == 'done':
                break
            if not choice.isdigit() or int(choice) < 1 or int(choice) > len(self.menu):
                print("Invalid choice. Try again.")
                continue
            item = self.menu[int(choice) - 1]
            order.append(item)
            print(f"Added {item.name} to the order.")
        if order:
            self.orders.append(order)
            print("Order placed successfully!")
        else:
            print("No items ordered.")

    def show_orders(self):
        if not self.orders:
            print("No orders yet.")
            return
        print("\n---- Orders ----")
        for i, order in enumerate(self.orders, 1):
            total = sum(item.price for item in order)
            item_names = ", ".join(item.name for item in order)
            print(f"Order {i}: {item_names} | Total: ₹{total}")
        print("-------\n")

# Function to display the main menu
def main():
    cafe = IraniCafe()
    # Predefined menu items
    cafe.add_menu_item("Irani Chai", 20)
    cafe.add_menu_item("Osmania Biscuit", 15)
    cafe.add_menu_item("Bun Maska", 25)
    cafe.add_menu_item("Keema Pav", 50)

    while True:
        print("\n Irani Cafe Management")
        print("1. Show Menu")
        print("2. Take Order")
        print("3. Show All Orders")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            cafe.show_menu()
        elif choice == '2':
            cafe.take_order()
        elif choice == '3':
            cafe.show_orders()
        elif choice == '4':
            print("Thank you for visiting the Irani Cafe! Visit Again.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the project
if __name__ == "__main__":
    main()
```
